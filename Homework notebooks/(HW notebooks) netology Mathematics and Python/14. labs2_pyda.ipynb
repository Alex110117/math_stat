## 6. Выдвинуть несколько гипотез и проверить при помощи t test

```python
df.columns
```

```python
atmo_city=df[df['aspiration']=='std']['city-mpg']
turbo_city=df[df['aspiration']=='turbo']['city-mpg']

sts.ttest_ind(atmo_city, turbo_city)
```

```python
print(len(atmo_city))
print(len(turbo_city))
```

```python
sns.distplot(atmo_city)
sns.distplot(turbo_city)
```

Нормальное распределение

```python
price_gas=df.loc[df['fuel-type']=='gas', 'price']
price_diesel=df.loc[df['fuel-type']=='diesel', 'price']

sts.ttest_ind(price_gas, price_diesel)
```

```python
sns.distplot(price_gas)
sns.distplot(price_diesel)
```

```python
len(price_gas), len(price_diesel)
```

Не нормальное распределение

```python
price_gas=price_gas.sample(n=30, replace=True)
price_diesel=price_diesel.sample(n=30, replace=True)
```

```python
sns.distplot(price_gas)
sns.distplot(price_diesel)
```

Распределение не нормализовалось, даже после взятия одинакового sample

```python
sts.ttest_ind(price_gas, price_diesel)
```
